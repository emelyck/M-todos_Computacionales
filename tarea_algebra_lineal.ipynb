```python
#punto 12 y punto 13
import numpy as np

def ecuaciones(x):
    p_1 = np.log(x[0]**2 + x[1]**2) - np.sin(x[0]*x[1]) - np.log(2 * np.pi)
    p_2 = np.exp(x[0] - x[1]) + np.cos(x[0]*x[1])
    return np.array([p_1, p_2])
def jacobiano(x):
    l_21 = np.exp(x[0] - x[1]) - (x[1] * np.sin(x[0]*x[1]))
    l_22 = -np.exp(x[0] - x[1]) - (x[0] * np.sin(x[0]*x[1]))
    l_11 = (2*x[0] / (x[0]**2 + x[1]**2)) - (x[1] * np.cos(x[0]*x[1]))
    l_12 = (2*x[1] / (x[0]**2 + x[1]**2)) - (x[0] * np.cos(x[0]*x[1]))
    return np.array([[l_11, l_12], [l_21, l_22]])
def newton_raphson(x0, max_iter=100, tol=1e-9):
    lx = np.copy(x0)
    for i in range(max_iter):
        J = jacobiano(lx)
        F = ecuaciones(lx)
        cambio_x = np.linalg.solve(J, -F)
        lx = lx + cambio_x
        if np.linalg.norm(cambio_x) < tol:
            break
    return lx
x_0 = np.array([2, 2])
s_n = newton_raphson(x_0)
print("Solución con N-R:", s_n)
def objetivo(x):
    p_1 = np.log(x[0]**2 + x[1]**2) - np.sin(x[0]*x[1]) - np.log(2 * np.pi)
    p_2 = np.exp(x[0] - x[1]) + np.cos(x[0]*x[1])
    return 0.5 * (p_1**2 + p_2**2)
def gradiente(x):
    F = ecuaciones(x)
    J = jacobiano(x)
    return np.dot(J.T, F)
def grad(x0, lr=0.01, tol=1e-9, max_iter=1000):
    x = np.copy(x0)
    for i in range(max_iter):
        grad = gradiente(x)
        x = x - lr * grad
        if np.linalg.norm(grad) < tol:
            break
    return x
solucion_gradiente = grad(x_0)
print("Solución gradiente:", solucion_gradiente)
def jacobiano(f, x, h=1e-2):
    n = len(x)
    J = np.zeros((n, n))
    for i in range(n):
        x_f_2 = np.copy(x)
        x_f_1 = np.copy(x)
        x_b_1 = np.copy(x)
        x_b_2 = np.copy(x)
        x_f_2[i] += 2*h
        x_f_1[i] += h
        x_b_1[i] -= h
        x_b_2[i] -= 2*h
        J[:, i] = (-f(x_f_2) + 8*f(x_f_1) - 8*f(x_b_1) + f(x_b_2)) / (12 * h)
    return
p_x = np.array([0.5, 0.5, 0.5])
def s_568(x):
    f1 = 6*x[0] - 2*np.cos(x[2]) - 1
    f2 = 9*x[1] + np.sqrt(x[0]**2 + np.sin(x[2])) + 0.9 + 1.06
    f3 = 60*x[2] + 3*np.exp(-x[0]*x[1]) + 10*np.pi - 3
    return np.array([f1, f2, f3])
J_4= jacobiano(s_568, p_x)
print("Jacobiano:")
print(J_4)
def segundo_orden(f, x, h=1e-2):
    n = len(x)
    k = np.zeros((n, n))
    for i in range(n):
        x_f = np.copy(x)
        x_b = np.copy(x)
        x_f[i] += h
        x_b[i] -= h
        k[:, i] = (f(x_f) - f(x_b)) / (2 * h)
    return k
s_o = segundo_orden(s_568, p_x)
print("segundo orden:")
print(s_o)
```

```python
#punto 14
import sympy as sp
x = sp.Matrix([[0, 0, 0], [0, 0, -1], [0, 1, 0]])
y = sp.Matrix([[0, 0, 1], [0, 0, 0], [-1, 0, 0]])
z = sp.Matrix([[0, -1, 0], [1, 0, 0], [0, 0, 0]])
def conmutador(A, B):
    return (A * B) - (B * A)
Jx_Jy = conmutador(x, y)
Jy_Jz = conmutador(y, z)
Jz_Jx = conmutador(z, x)
print("Jx, Jy")
sp.pprint(Jx_Jy)
print("Jy, Jz:")
sp.pprint(Jy_Jz)
print("Jz, Jx:")
sp.pprint(Jz_Jx)
print("Jx, Jy= Jz")
sp.pprint(Jx_Jy.equals(z))
print("Jy, Jz = Jx")
sp.pprint(Jy_Jz.equals(x))
print("Jz, Jx = Jy")
sp.pprint(Jz_Jx.equals(y))
```
