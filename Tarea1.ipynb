```python
#solición del ejercicio Capitulo 2.0.2 Máximos
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import urllib.request
link = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
info = urllib.request.urlopen(link)
ejesdatos= np.loadtxt(info)
ejex_datos = ejesdatos[:, 0]
ejey_datos = ejesdatos[:, 1]
peaks, _ = find_peaks(ejey_datos)
plt.plot(ejex_datos, ejey_datos)
plt.plot(ejex_datos[peaks], ejey_datos[peaks],'ro',)
plt.show()
#para este ejercicio hicimos uso de librerías numpy, matplotib y scipy, este ultimo nos proporciona funciones que buscan los máximos""
```

```python
#solición del ejercicio Capitulo 2.0.5 Espiral de arquímedes.
import numpy as np
import matplotlib.pyplot as plt
a = 0
b = 1
theta = np.linspace(0, 2 * np.pi, 100)
r = a + b * theta
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.plot(x, y)
plt.title('Espiral de Arquímedes')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
#se hizo uso de las librerías numpy y matplotib
```

```python
#solición del ejercicio Capitulo 4- punto 6
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 25)
y = np.linspace(-4, 4, 25)
X, Y = np.meshgrid(x, y)
Radio = 2
Velocidad = 2

fun_pot_flujo = Velocidad * X * (1 - (Radio**2 / (X**2 + Y**2)))
h = 0.001
Vx_parcial = (np.roll(fun_pot_flujo, -1, axis=1) - np.roll(fun_pot_flujo, 1, axis=1)) / (2*h)
Vy_parcial = (np.roll(fun_pot_flujo, -1, axis=0) - np.roll(fun_pot_flujo, 1, axis=0)) / (2*h)


interior = X**2 + Y**2 < Radio**2
Vx_parcial[interior] = 0
Vy_parcial[interior] = 0

plt.quiver(X, Y, Vx_parcial, Vy_parcial, color='b')
circle = plt.Circle((0, 0), Radio, fill=False)
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.xlabel('x')
plt.ylabel('y ')
plt.gca().add_patch(circle)
plt.show()
```
