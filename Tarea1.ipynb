```python
#solición del ejercicio Capitulo 2.0.2 Máximos
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import urllib.request
link = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'
info = urllib.request.urlopen(link)
ejesdatos= np.loadtxt(info)
ejex_datos = ejesdatos[:, 0]
ejey_datos = ejesdatos[:, 1]
peaks, _ = find_peaks(ejey_datos)
plt.plot(ejex_datos, ejey_datos)
plt.plot(ejex_datos[peaks], ejey_datos[peaks],'ro',)
plt.show()
#para este ejercicio hicimos uso de librerías numpy, matplotib y scipy, este ultimo nos proporciona funciones que buscan los máximos""
```
