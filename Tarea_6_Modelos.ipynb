```python
import numpy as np
import matplotlib.pyplot as plt

#Metodo de MonteCarlo
radio_1 = 1
pun_gene = 10000
den = 1
ang = np.random.uniform(0, 2 * np.pi, pun_gene)
polar = np.random.uniform(0, np.pi, pun_gene)
radio_2 = radio_1 * (np.random.rand(pun_gene)) ** (1/3)
x = radio_2 * np.sin(polar) * np.cos(ang)
y = radio_2 * np.sin(polar) * np.sin(ang)
z = radio_2 * np.cos(polar)
Ixx = np.mean(y**2 + z**2)
Iyy = np.mean(x**2 + z**2)
Izz = np.mean(x**2 + y**2)
distri = -np.mean(x * y)
print(f"Ixx = {Ixx:.4f}")
print(f"Iyy = {Iyy:.4f}")
print(f"Izz = {Izz:.4f}")
print(f"Ixy = {distri:.4f}")
print(f" Ixx, Iyy, Izz: 0.4 ")
fig = plt.figure(figsize=(11, 8))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(x, y, z, s=1)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(x, y, z, s=1)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")
plt.tight_layout()
plt.show()
```
