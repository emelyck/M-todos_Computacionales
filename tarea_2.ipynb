```python
#punto de Derivación 8, Punto C
import numpy as np
import matplotlib.pyplot as plt
def funcion(x):
    return np.sqrt(np.tan(x))
h = 0.01
valores_de_x = np.arange(0.1, 1.1 + h, h)
def progresiva(f, x, h):
    return (1 / (2 * h)) * (-3 * f(x) + 4 * f(x + h) - f(x + 2 * h))

f_inter = progresiva(funcion, valores_de_x, h)



#punto de Derivación 8, Punto D
def central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
f_inter2 = central(funcion, valores_de_x, h)









```
