```python
#punto de Derivación 8, Punto C
import numpy as np
import matplotlib.pyplot as plt
def funcion(x):
    return np.sqrt(np.tan(x))
h = 0.01
valores_de_x = np.arange(0.1, 1.1 + h, h)
def progresiva(f, x, h):
    return (1 / (2 * h)) * (-3 * f(x) + 4 * f(x + h) - f(x + 2 * h))

f_inter = progresiva(funcion, valores_de_x, h)


#punto de Derivación 8, Punto D
def central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
f_inter2 = central(funcion, valores_de_x, h)


#punto de Derivación 8, Punto e parte de grafica hecha con código
def f_normal(x):
    return (1 / (2 * np.sqrt(np.tan(x)))) * (1 / np.cos(x)**2)

f_datos = f_normal(valores_de_x)

plt.plot(valores_de_x, f_datos, label=' Analítica')
plt.plot(valores_de_x, f_inter, '--', label='Progresiva')
plt.plot(valores_de_x, f_inter2, '--', label='Central')
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('punto e')
plt.legend()
plt.grid(True)

#punto de Derivación 8, Punto f parte de grafica hecha con código

plt.figure(figsize=(10, 6))
error_p = np.abs(f_inter - f_datos)
error_c = np.abs(f_inter2 - f_datos)
plt.plot(valores_de_x, error_p, label='Error P')
plt.plot(valores_de_x, error_c, label='Error C')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Punto f')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()




```

```python
#interpolación de Lagrange punto 4

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
x = np.array([1.4, 3.5, 5.6])
y = np.array([0.4007954931819738, 0.594128102489774, 0.29802795523938164])
def cuadratica(x, a, b, c):
    return a * x**2 + b * x + c
coeficientes, cov = curve_fit(cuadratica, x, y)
a, b, c = coeficientes
g = 9.8
angulo = np.arctan(b)
angulo_en_grados = np.degrees(angulo)
V0 = np.sqrt(-g / (2 * a * np.cos(angulo)**2))
print(f" V0 = {V0:.2f} m/s")
print(f"θ = {angulo_en_grados:.2f} grados")
print(f"coeficientes de la función de ajuste cuadratico a = {a}, b = {b}, c = {c}")
x_la = np.linspace(min(x), max(x), 50)
plt.scatter(x, y)
plt.plot(x_la, cuadratica(x_la, *coeficientes))
plt.xlabel('x')
plt.ylabel('y ')
plt.title('púnto 4, trayectoria que hace la bala')
plt.legend()
plt.show()
```
